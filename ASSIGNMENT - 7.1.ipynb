## Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime = pd.read_csv(r"C:\Users\SUMAN\Desktop\ASSIGNMENT- EXCELR\crime_data.csv")
crime
```

```python
crime1 = crime.rename({'Unnamed: 0' : 'Place'},axis=1)
crime1 = crime1[['Place','Murder','Assault','UrbanPop','Rape']]
```

```python
#normalize the data
def norm_fun(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df1=norm_fun(crime1.iloc[:,1:])
```

```python
numeric_columns = crime1.select_dtypes(include=[np.number])
```

```python
numeric_columns
```

```python
#Elbow cruve
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(numeric_columns)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title("Elbow curve")
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model = KMeans(n_clusters=4)
model.fit(numeric_columns)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2 = pd.read_csv(r"C:\Users\SUMAN\Desktop\ASSIGNMENT- EXCELR\crime_data.csv")
crime2
```

```python
data2 = crime2.rename({'Unnamed: 0' : 'Place'},axis=1)
```

```python
data2.info()
```

```python
data2.describe
```

```python
df2 = data2.iloc[:,1:5]
df2
```

```python
df2.values
```

```python
stscaler=StandardScaler().fit(df2.values)
x=stscaler.transform(df2.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
clu=pd.DataFrame(dbscan.labels_,columns=['cluster'])
clu
```

```python
pd.concat([clu,data2],axis=1)
```
